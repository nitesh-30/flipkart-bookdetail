```python
import requests
content=requests.get("https://www.flipkart.com/search?q=book&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
content
```

```python
data = bs(content.content,"html.parser")
```

```python
name=data.find_all('a',class_="s1Q9rs")
Book_name=[]
for p in name:
    Book_name.append(p.text)
```

```python
img=data.find_all('img',class_="_396cs4 _3exPp9")
Book_img=[]
for i in img:
    Book_img.append(i.get('src'))
```

```python
p=data.find_all('div',class_="_30jeq3")
Price=[]
for j in p:
    Price.append(j.text)
```

```python
r=data.find_all('div', class_="_3LWZlK")

rating=[]
for ra in r:
    rating.append(ra.text)
```

```python
Price=Price[0:40]
rating=rating[0:40]
```

```python
df={'BOOK_NAME':Book_name,'BOOK_PRICE':Price,'Book rating':rating,'Bookimage':Book_img}
```

```python
new=pd.DataFrame.from_dict(df)
```
